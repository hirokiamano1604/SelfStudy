# Limits of Diversification

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import edhec_risk_kit as erk
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ind_return = erk.get_ind_returns()
ind_nfirms = erk.get_ind_nfirms()
ind_size = erk.get_ind_size()
```

## nfirm CSV

## size CSV

#Variables

 - ind_return is the averaged returns of the industry
 - ind_nfirms is the number of the companies in the industry
 - ind_size is the averaged market capital of the companies in the industry
 - ind_mktcap is the sum of the market capital of the companies in the industry
 - total_mktcap


```python
ind_return.shape
```

```python
ind_nfirms.shape
```

```python
ind_mktcap = ind_nfirms * ind_size
```

```python
total_mktcap = ind_mktcap.sum(axis = "columns")
```

```python
total_mktcap.plot(figsize = (12,6))
plt.show()
```

```python
ind_capweight = ind_mktcap.divide(total_mktcap, axis = "index")
```

```python
ind_capweight
```

```python
ind_capweight.loc["1926"].sum(axis ="columns")
```

```python
ind_capweight[["Fin","Steel"]].plot(figsize=(12,6))
plt.show()
```

```python
total_market_returns = (ind_capweight * ind_return).sum(axis = "columns")
```

```python
total_market_returns.plot(figsize = (12,6))
plt.show()
```

```python
total_market_index = erk.drawdown(total_market_returns).Wealth
total_market_index.plot(figsize = (10,5), title = "Total MArket CapWeighted Index")
plt.show()
```

```python
total_market_index["1980":].plot(figsize = (12,6))
total_market_index["1980":].rolling(window=36).mean().plot(figsize = (12,6))
plt.show()
```

```python
tmi_tr36rets = total_market_returns.rolling(window=36).aggregate(erk.annualize_rets,periods_per_year=12)
tmi_tr36rets.plot(figsize=(12,6), label="Tr36 month returns",legend=True)
total_market_returns.plot(label="Returns",legend=True)
plt.show()
```

# Rolling Correlation - sling with Muti Indexes and .groupby

```python
ts_corr = ind_return.rolling(window=36).corr()
ts_corr.tail()
```

```python
ts_corr.index.names = ['date','industry']
ts_corr.tail()
```

```python
ind_tr36corr = ts_corr.groupby(level="date").apply(lambda covmat: covmat.values.mean())
ind_tr36corr.plot()
plt.show()
```

```python
tmi_tr36rets.plot(label="Tr36 Month Rets", figsize=(12,6),legend=True)
ind_tr36corr.plot(label="Tr36 Month Corr", legend = True, secondary_y=True)
plt.show()
```

```python
tmi_tr36rets["2007":].plot(label="Tr36 Month Rets", figsize=(12,6),legend=True)
ind_tr36corr["2007":].plot(label="Tr36 Month Corr", legend = True, secondary_y=True)
plt.show()
```
